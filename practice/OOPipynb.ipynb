```python
class AreaCal:
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth

  @property
  def find_area(self):
    return self.length * self.breadth

  @find_area.setter
  def find_area(self, new_area):
    self.length = new_area/self.breadth
```

```python
class Students:
  def __init__(self, marks):
    self.marks = marks

  @property
  def student_grader(self):
    return self.marks

  @student_grader.setter
  def marks_grades(self,value):
    if value > 100 or value < 0:
      print(f"invalid marks")
    else:
      return self.marks

  @property
  def grader(self):
    if self.marks >= 90:
      return "A"

    elif self.marks >=75:
      return "B"

    elif self.marks >=50:
      return "c"



```

```python
final_grader = Students(55)
```

```python
print(final_grader.grader)
```
