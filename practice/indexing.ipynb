import numpy as np

def eu_sim(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.sqrt(np.sum((a - b) ** 2))

def cos_sim(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))


class FlatIndex:
    def __init__(self):
        self.text = []
        self.vectors = []

    def add_vectors(self, vector, text):
        self.text.append(text)
        self.vectors.append(np.array(vector))

    def search_query(self, search_query, top_k=1, metric="cosine"):
        search_query = np.array(search_query)
        scores = []

        for i, vector in enumerate(self.vectors):
            if metric == "cosine":
                score = cos_sim(search_query, vector)
            elif metric == "euclidean":
                score = eu_sim(search_query, vector)
            else:
                raise ValueError("metric must be 'cosine' or 'euclidean'")

            scores.append((score, self.text[i]))

        if metric == "cosine":
            scores.sort(reverse=True, key=lambda x: x[0])   
        else:
            scores.sort(key=lambda x: x[0])                 

        return scores[:top_k]


```python
import numpy as np

def euclid_distance(a,b):
    a = np.array(a)
    b = np.array(b)

    eu_score = np.sqrt(np.sum((a-b)**2))
    return eu_score
```
