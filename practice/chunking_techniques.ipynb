```python
def para_chunking(text, max_chunk_size = 100):
    paragraphs = text.split("\n\n")
    chunk = []
    current_chunk = ""

    for paragraph in paragraphs:
        paragraph = paragraph.strip()
        if not paragraph:
            continue
        if len(paragraph)+len(current_chunk)<=max_chunk_size:
            current_chunk += paragraph + "\n\n"
        else:
            chunk.append(current_chunk.strip())
            current_chunk = paragraph + "\n\n"
    if current_chunk:
            chunk.append(current_chunk.strip())
    return chunk
            
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer("all-miniLM-L6-v2")

def sim_score(a,b):
    score = np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
    return score

def semantic_chunk(text, threshold = 0.7):
    sentences = [s.strip() for s in text.split(".") if s.strip()]
    chunks = []
    current_chunk = ""

    for i in range(1, len(sentences)):
        emb1 = model.encode(current_chunk[-1])
        emb2 = model.encode(sentences[i])
        sim = sim_score(emb1, emb2)

    if sim >= threshold:
        current_chunk.append(sentences[i])
    
    else:
        chunks.append(". ".join(current_chunk) + ". ")
        current_chunk = [sentences[i]]

    if current_chunk:
        chunks.append(". ".join(current_chunk) + ". ")
    return chunks

    
```

```python
sentences = [
    "Python is a popular programming language",
    "It is widely used in data science",
    "It has many libraries like NumPy and Pandas",
    "Cricket is a very popular sport in India",
    "Many people watch cricket matches",
    "Data science involves statistics and machine learning",
    "It is used to analyze data"
]
```

```python
similarity = {
    (0,1): 0.88,
    (1,2): 0.82,
    (2,3): 0.25,
    (3,4): 0.90,
    (4,5): 0.30,
    (5,6): 0.85
}
```

```python
def semantic_chunking(text, threshold =.75):
    sentence = [s.strip() for s in text.split(".") if s.strip]
    chunk = []
    last_chunk = [sentence[0]]

    for i in range(1, len(sentence)):
        if similarity [(i-1), i] >= threshold:
            last_chunk.append(sentences[i])
            
        else:
            chunk.append(". ".join(last_chunk[-1]) + ". ")
            last_chunk = [sentences[i]]
        if last_chunk:
            chunk.append(". ".join(last_chunk) + ". ")
        return chunk
```
